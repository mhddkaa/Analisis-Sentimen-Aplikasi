```python
from google_play_scraper import Sort, reviews
from google_play_scraper import app
import pandas as pd
import numpy as np
```

```python
result, continuation_token = reviews(
    'com.djinggamedia.denpasarsightseeing',
    lang='id', # Bahasa
    country='id', # Negara
    sort=Sort.MOST_RELEVANT, # Sorting yang paling relevan
    count=500, # Jumlah dataset
    filter_score_with= None
)
```

```python
# Dataframe
data = pd.DataFrame(np.array(result),columns=['review'])
data = data.join(pd.DataFrame(data.pop('review').tolist()))
data.head()
```

```python
len(data)
```

```python
data = data[['content','score']]
data.head()
```

```python
data = data.rename(columns={'content': 'ulasan', 'score': 'rating'}) # Mengganti nama fitur
data.head()
```

```python
# Labeling sentimen berdasarkan rating
def label_sentiment(rating):
    if rating in [1, 2]:
        return 'Negatif'
    elif rating == 3:
        return 'Netral'
    elif rating in [4, 5]:
        return 'Positif'

# Terapkan ke kolom rating
data['sentiment'] = data['rating'].apply(label_sentiment)

# Melihat hasil
print(data[['rating', 'sentiment']].head())
```

```python
data.to_csv("UlasanDPS.csv", index = False  , encoding='utf-8')
```
